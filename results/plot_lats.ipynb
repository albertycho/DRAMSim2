```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
lats = []
sum = 0
max_val = 0
lat_bins = [0]
with open("k6_aoe_02_short.trc/DDR3_micron_64M_8B_x4_sg15/4GB.1Ch.1R.scheme2.open_page.32TQ.32CQ.RtB.pRank.vis") as f:
    [f.readline() for _ in range(65)]
    for line in f.readlines()[:-1]:
        val, num = line.split("=")
        sum += int(num)
        max_val = max(int(val), max_val)
        
        lat_bins.append(int(val))
        lats.append(int(num))

sum, max_val
```

```python
prio_lats = []
sum = 0
max_val = 0
prio_bins = [0]
with open("k6_aoe_02_short.trc/DDR3_micron_64M_8B_x4_sg15/prio.4GB.1Ch.1R.scheme2.open_page.32TQ.32CQ.RtB.pRank.vis") as f:
    [f.readline() for _ in range(65)]
    for line in f.readlines()[:-1]:
        val, num = line.split("=")
        sum += int(num)
        max_val = max(int(val), max_val)
        
        prio_bins.append(int(val))
        prio_lats.append(int(num))

sum, max_val
```

```python
# Plot lats

# Creating histogram
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
 
axs.hist(lats, bins = lat_bins)
 
# Show plot
plt.show()
```

```python
# Plot prio lats

# Creating histogram
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
 
axs.hist(prio_lats, bins = prio_bins)
 
# Show plot
plt.show()
```
