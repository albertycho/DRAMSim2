```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
lats = []
sum = 0
max_val = 0
lat_bins = [0]
prio_only_lats = []
prio_only_lat_bins = []
with open("k6_aoe_02_short.trc/DDR3_micron_64M_8B_x4_sg15/test_4GB.1Ch.1R.scheme2.open_page.32TQ.32CQ.RtB.pRank.vis") as f:
    [f.readline() for _ in range(65)]
    prio = False
    for line in f.readlines()[:-1]:
        if not prio:
            if line == "!!PRIO_HISTOGRAM_DATA":
                prio = True
                continue
            # try:
            val, num = line.split("=")
            # except ValueError as e:
            #     print(line)
            sum += int(num)
            max_val = max(int(val), max_val)
            
            lat_bins.append(int(val))
            lats.append(int(num))
        else:
            val, num = line.split("=")
            
            prio_only_lat_bins.append(int(val))
            prio_only_lats.append(int(num))
    
sum, max_val
```

```python
# Plot lats

# Creating histogram
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
 
axs.hist(lats, bins = lat_bins)
 
# Show plot
plt.show()
```

```python
# Plot lats

# Creating histogram
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
 
axs.hist(prio_only_lats, bins = prio_only_lat_bins)
 
# Show plot
plt.show()
```
